```python
import numpy as np
from typing import Tuple
from matplotlib import pyplot as plt
import random
```

```python
def f_1(x):
    x = np.float64(x)
    temp = x ** 4
    temp *= np.e ** x
    temp -= np.sin(x)
    temp /= 2
    return temp
```

```python
def f_2(x):
    x = np.float64(x)
    temp = 5 * x 
    temp = np.sin(temp)
    temp += np.sqrt(x)
    temp = 5 * np.log(temp)
    return temp
```

```python
def f_3(x):
    x = np.float64(x)
    temp = 5 * np.log10(x)
    temp += np.cos(x)
    temp -= (x ** 3) / 10
    return temp
```

ب) برای تابع سوم (تابع مقعر) می‌توان با استفاده از الگوریتمی مانند کاهش گرادیان کمینه تابع
f(x)-
را یافت که متناظر با بیشینه خود تابع است. برای تابع دوم روشی که که تضمین یافتن نقطه فرینه سراسری را بدهد موجود نیست.

```python
def gradiant_descent(func, initial_point: float, learning_rate: float, max_iterations: int, func_range: tuple):
    x = []
    y = []
    iter = 0
    epsilon = 10e-6
    min_x = initial_point
    while iter < max_iterations:
        x.append(min_x)
        y.append(func(min_x))
        grad = (func(min_x + epsilon) - func(min_x)) / (epsilon)
        if grad == 0:
            break
        min_x -= learning_rate * grad
        min_x = min(min_x, func_range[1])
        min_x = max(min_x, func_range[0])
        iter += 1
    return min_x, np.float32(x), np.float32(y)
```

```python
min_x, x, y = gradiant_descent(
    func=f_1, 
    initial_point=np.random.uniform(-2, 1), 
    learning_rate=0.1, 
    max_iterations=1000,
    func_range=(-2, 1)
)
plt.plot(x,y, 'b')
plt.show()
```

```python
def nr(func, initial_point: float, max_iterations: int, func_range: tuple):
    x = []
    y = []
    iter = 0
    epsilon = 10e-6
    min_x = initial_point
    while iter < max_iterations:
        x.append(min_x)
        y.append(func(min_x))
        grad = (func(min_x + epsilon) - func(min_x)) / (epsilon)
        lr = (epsilon ** 2) / (func(min_x + epsilon) + func(min_x - epsilon) - 2 * func(min_x))
        min_x -= lr * grad
        if grad == 0:
            break
        min_x = min(min_x, func_range[1])
        min_x = max(min_x, func_range[0])
        iter += 1
    return min_x, np.float32(x), np.float32(y)
```

```python
def f(x_1, x_2):
    temp = (2 ** x_1) / 10000
    temp += (np.e ** x_2) / 20000
    temp += x_1 ** 2
    temp += 4 * (x_2 ** 2)
    temp -= 2 * x_1
    temp -= 3 * x_2
    return temp
```

```python
def gradiant_descent(func, initial_point: Tuple, learning_rate: float, threshold: float, max_iterations: int):
    x_1_sequence = [initial_point[0]]
    x_2_sequence = [initial_point[1]]
    iter = 0
    while iter < max_iterations and x_1_sequence[-1] <= threshold and x_2_sequence[-1] <= threshold:
        x_1_temp, x_2_temp = update_points(func, x_1_sequence[-1], x_2_sequence[-1], learning_rate)
        x_1_sequence.append(x_1_temp)
        x_2_sequence.append(x_2_temp)
        iter += 1
    return x_1_sequence, x_2_sequence

def update_points(func, x_1, x_2, learning_rate):
    eps = 10e-6
    grad_x_1 = (func(x_1 + eps, x_2) - func(x_1, x_2)) / eps
    grad_x_2 = (func(x_1, x_2 + eps) - func(x_1, x_2)) / eps
    x_1 -= learning_rate * grad_x_1
    x_2 -= learning_rate * grad_x_2
    return x_1, x_2
```

```python
def draw_points(func, x_1_sequence, x_2_sequence):
    fig = plt.figure(figsize=plt.figaspect(0.5))
    X1, X2 = np.meshgrid(np.linspace(-15.0, 15.0, 1000), np.linspace(-15.0, 15.0, 1000))
    Y = func(X1, X2)
    f_sequence = [f(x_1_sequence[i], x_2_sequence[i]) for i in range(len(x_1_sequence))]

    # First subplot
    ax = fig.add_subplot(1, 2, 1)

    cp = ax.contour(X1, X2, Y, colors='black', linestyles='dashed', linewidths=1)
    ax.clabel(cp, inline=1, fontsize=10)
    cp = ax.contourf(X1, X2, Y, )
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.scatter(x_1_sequence, x_2_sequence, s=10, c="y")

    # Second subplot
    ax = fig.add_subplot(1, 2, 2, projection='3d')

    ax.contour3D(X1, X2, Y, 50, cmap="Blues")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.scatter3D(x_1_sequence, x_2_sequence, f_sequence, s=10, c="r")

    plt.show()
```

```python
initial_point = (np.random.uniform(-15, 15), np.random.uniform(-15, 15))
learning_rate = 0.01
threshold = 100
max_iterations = 1000
x_1_sequence, x_2_sequence = gradiant_descent(func=f, initial_point=initial_point, learning_rate=learning_rate, threshold=threshold, max_iterations=max_iterations)
print((x_1_sequence[-1], x_2_sequence[-1]))
draw_points(f, x_1_sequence, x_2_sequence)
```

```python
def random_state_generator(func_range):
    return np.random.uniform(func_range[0], func_range[1])

def neighbour_state_generator(x, alpha, func_range):
    left = max(func_range[0], x - alpha)
    right = min(func_range[1], x + alpha)
    return np.random.uniform(left, right)

def cost_function(func, x):
    return -func(x)

def prob_accept(func, current_x, next_x, T):
    delta_F = cost_function(func, next_x) - cost_function(func, current_x)
    return np.exp(delta_F / T )

def accept(func, current_x, next_x, T):
    delta_F = cost_function(func, next_x) - cost_function(func, current_x)
    if delta_F > 0:
        return True
    accept_prob = prob_accept(func, current_x, next_x, T)
    return random.choices([True, False], weights=(accept_prob, 1 - accept_prob))

def anneal(
    func, func_range, alpha, stopping_temperature=1e-8 , stopping_iter=2000, gamma=0.99, T=50):
    counter = 0
    current_state = random_state_generator(func_range)
    while T > stopping_temperature:
        if counter == stopping_iter:
            break
        T = gamma * T
        next_state = neighbour_state_generator(current_state, alpha, func_range)
        if accept(func, current_state, next_state, T):
            current_state = next_state
    best_solution = current_state
    return best_solution
```

```python
min_x_of_f2 = 2.18
eps = 0.1
for i in range(1, 11):
    counter = 0
    for j in range(1000):
        min_x = anneal(f_2, (2, 6), i / 10)
        counter += (np.abs(min_x - min_x_of_f2) < eps)
    print('success percent for alpha = ', i / 10, ' is: ', round((counter / 1000) * 100, 2), '%')
    
```
